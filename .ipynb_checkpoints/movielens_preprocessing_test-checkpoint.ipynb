#### <span style="color:#993399">Créer la fct générique (qui fait tout le travail ci-dessus) qui regroupe les éléments(eg. movies) par utilisateurs</span>

```python
# Créer la fct générique (qui fait tout le travail ci-dessus) qui regroupe les éléments(eg. movies) par utilisateurs :
def group_items(df, grouping_col, grouped_col):
    grouped_df = df.groupby(grouping_col)[grouped_col].apply(list).to_frame()
    return grouped_df
```

```python
rating_grouped = group_items(df = rating, grouping_col = "userId", grouped_col = "movieId")
rating_grouped
```
